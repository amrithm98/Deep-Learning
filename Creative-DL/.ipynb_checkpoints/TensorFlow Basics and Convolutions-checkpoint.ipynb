```python
sess = tf.InteractiveSession()
```

```python
#uses interactive session
seq_tf = tf.linspace(-3.0,3.0,100)
seq_tf.eval()
```

```python
opList = [op.name for op in currGraph.get_operations()]
print(opList)
```

```python
seq_tf.get_shape().as_list()
```

```python
mean = 0.0
sigma = 1.0
```

```python
normal = (tf.exp(tf.negative(tf.pow(seq_tf - mean, 2.0) /
                   (2.0 * tf.pow(sigma, 2.0)))) *
     (1.0 / (sigma * tf.sqrt(2.0 * 3.1415))))
```

```python
res = normal.eval()
```

```python
#Normal Distribution
plt.plot(res)
```

## Gaussian Kernel, Gabor Kernel & Convolutions

```python
kernelSize = normal.get_shape().as_list()[0]
print(kernelSize)
```

```python
M1 = tf.reshape(normal,[kernelSize,1])
M2 = tf.reshape(normal,[1,kernelSize])
```

```python
gaussKernel = tf.matmul(M1,M2)
kernel = gaussKernel.eval()
plt.imshow(kernel)
```

```python
kernel.shape
```

```python
from skimage import data
```

```python
img = data.camera().astype(np.float32)
```

```python
plt.imshow(img,cmap='gray')
```

```python
print(kernel)
```

```python
image_4d = tf.reshape(img,[1,img.shape[0],img.shape[1],1])
```

```python
kernel_4d = tf.reshape(kernel,[kernel.shape[0],kernel.shape[1],1,1])
```

```python
print(kernel_4d)
```

```python
convolution = tf.nn.conv2d(image_4d,kernel_4d,[1,1,1,1],'SAME')
```

```python
convolved_image = convolution.eval()
```

```python
print(convolved_image.shape)
```

```python
#We blurred the image using Gaussian Kernel
plt.imshow(convolved_image.reshape(512,512))
```

```python
plt.imshow(convolved_image[0,:,:,0],cmap='gray')
```

## Gabor Kernel
Modulating Gaussian Kernel using a sine wave

```python
xs = tf.linspace(-3.0,3.0,100)
ys = tf.sin(xs)
```

```python
plt.plot(ys.eval())
```

```python
M1 = tf.reshape(ys,[kernelSize,1])
M2 = tf.ones([1,kernelSize])
wave = tf.matmul(M1,M2)
```

```python
plt.imshow(wave.eval(),cmap='gray')
```

```python
gabor = tf.matmul(wave,kernel)
```

```python
plt.imshow(gabor.eval(),cmap='gray')
```

```python
conv_2 = tf.nn.conv2d(image_4d,tf.reshape(gabor,[kernelSize,kernelSize,1,1]).eval(),[1,1,1,1],'SAME')
```

```python
convolved_2 = conv_2.eval()
```

```python
plt.imshow(convolved_2[0,:,:,0],cmap='gray')
```

## Convolution Using Place Holders

```python
img = tf.placeholder(tf.float32,shape=[None,None],name = 'img')
```

```python
img_3d = tf.expand_dims(img,2)
```

```python
print(img_3d.get_shape().as_list())
```

```python
img_4d = tf.expand_dims(img_3d,0)
```

```python
print(img_4d.get_shape().as_list())
```

```python
mean = tf.placeholder(tf.float32,name='mean')
sigma = tf.placeholder(tf.float32,name='sigma')
kernelSize = tf.placeholder(tf.int32,name='kernelSize')
```

```python
x = tf.linspace(-3.0,3.0,kernelSize)
normal = (tf.exp(tf.negative(tf.pow(x - mean, 2.0) /
                   (2.0 * tf.pow(sigma, 2.0)))) *
     (1.0 / (sigma * tf.sqrt(2.0 * 3.1415))))
```

```python
M1 = tf.reshape(normal,[kernelSize,1])
M2 = tf.reshape(normal,[1,kernelSize])
gaussKernel = tf.matmul(M1,M2)
```

```python
ys = tf.sin(x)
```

```python
ys = tf.reshape(ys,[kernelSize,1])
```

```python
ones = tf.ones([1,kernelSize])
```

```python
wave = tf.matmul(ys,ones)
```

```python
gabor = tf.matmul(wave,gaussKernel)
```

```python
gabor_4d = tf.reshape(gabor,shape = [kernelSize,kernelSize,1,1])
```

```python
convolved_image = tf.nn.conv2d(img_4d,gabor_4d,[1,1,1,1],'SAME')
```

```python
#Use of Using PlaceHolders
conved = convolved_image.eval(feed_dict = {img : data.camera(), mean:0.0,sigma:1.0,kernelSize:100})
```

```python
plt.imshow(conved[0,:,:,0],cmap='gray')
```

```python
#Use of Using PlaceHolders
conved_2 = convolved_image.eval(feed_dict = {img : data.camera(), mean:0.0,sigma:1.0,kernelSize:10})
plt.imshow(conved_2[0,:,:,0],cmap='gray')
```
